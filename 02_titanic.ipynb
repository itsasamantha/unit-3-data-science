# Titanic Tabulation

The Titanic is a ship that set sail in 1912 and sank a few days after its departure. From April 10, 1912 to April 15, 1912, the Titanic was sailig for a mere 5 days. This notebook analyzes questions that arise from the 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('titanic_passengers.csv')
```

## How many passengers survived.

```python
print(titanic_df['Survived'].value_counts())
survived_count = titanic_df['Survived'].value_counts()

labels = 'Died', 'Survived'
sizes = survived_count

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors = ['LightSteelBlue', 'LightSlateGrey'], shadow=True, startangle=80)
```

As seen in the pie chart above 38.4% of the titanic passengers survived. This percent is 342 people out of the 891 total people.

# How many people embarked from Queenstown


```python
print(titanic_df['Embarked'].value_counts())
embarked_place = titanic_df['Embarked'].value_counts()

labels = 'Southampton', 'Cherbourg','Queenstown'
sizes = embarked_place

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors = ['LightSteelBlue', 'LightSlateGrey','SlateGrey'], shadow=True, startangle=80)
```

This graph shows the percentages of people embarking from each point. It shows that the least ,77/889 (two were not recorded), people embarked from Queenstown.

# Were passengers who paid for expensive tickets( over 50 pounds) more likely to survive?.

```python
expensive_fares = titanic_df[titanic_df['Fare'] > 50]
expensive_survival = expensive_fares['Survived'].value_counts()

cheap_fares = titanic_df[titanic_df['Fare'] <= 50]
cheap_survival = cheap_fares['Survived'].value_counts()

print(expensive_survival)
print(cheap_survival)

plt.figure(figsize=(10, 6))
# importing package 
import matplotlib.pyplot as plt 

# create data 
x = ["Survived","Died"]
y1 = expensive_survival
y2 = cheap_survival
width = 0.4

# plot data in grouped manner of bar type 
plt.bar("HI", y1, width) 
plt.bar("ce", y2, width) 

```
